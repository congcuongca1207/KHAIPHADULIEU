## 4. Hàm tổng hợp (Aggregation Function)

**13 hàm tổng hợp in Pandas**

`mean()`: Tính giá trị trung bình.

`sum()`: Tính tổng của các giá trị.

`size()`: Đếm số lượng phần tử.

`count()`: Đếm số lượng giá trị không rỗng.

`std()`: Tính độ lệch chuẩn.

`var()`: Tính phương sai.

`sem()`: Tính độ lệch chuẩn của trung bình.

`describe()`: Tạo ra các thống kê mô tả.

`first()`: Lấy giá trị đầu tiên

`last()`: Lấy giá trị cuối cùng.

`nth()`: Lấy giá trị thứ n hoặc một tập hợp con nếu n là một danh sách.

`min()`: Tìm giá trị nhỏ nhất.

`max()`: Tìm giá trị lớn nhất.

## 5. Lọc dữ liệu

````
df.head()
````

````
# Chọn dữ liệu có birth rate > 20
df[df['Birth rate'] > 20]
````

````
# Chọn dữ liệu có Internet users < 40
# YOUR CODE HERE
df[df['Internet users'] < 40]
````

***So sánh trong Python:***

```
bằng (equal): ==

khác (different): !=

lớn hơn (more than): >

nhỏ hơn (less than): <

lớn hơn hoặc bằng (more than or equal): >=

nhỏ hơn hoặc bằng (less than or equal): <=
```

````
# Ví dụ: tính birth rate trung bình của tất cả quốc gia trong nhóm high income
df[df['Income Group'] == 'High income']['Birth rate'].mean()
````

````
# Lọc dữ liệu Birth rate > 20 và Internet users < 50
# 1/ và (and): &
# 2/ hoặc (or) : |
# 3/ bọc từng điều kiện vào trong ngoặt tròn
df[(df['Birth rate'] > 20) | (df['Internet users'] < 50)]
````

````
# Lọc dữ liệu dựa trên nhiều điều kiện
# Nhớ có ngoặc () cho mỗi điều kiện
df[(df['Birth rate'] > 20) & (df['Internet users'] < 50)]
df[(df['Birth rate'] > 20) | (df['Internet users'] < 50)]
````

## 6. Sắp xếp dữ liệu

````
# Sắp xếp dữ liệu theo internet users từ lớn nhất từ nhỏ nhất (theo thứ tự giảm dần)
df = df.sort_values('Internet users', ascending=False)
df
````

````
# Tìm top 5 quốc gia theo tỷ lệ sinh birthrate
# YOUR CODE HERE
df.sort_values('Birth rate', ascending=False).head()
````
